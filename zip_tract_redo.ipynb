```python
# importing the HUD crosswalk tract to zip data

tractzip16 = pd.read_excel('/Users/k.simmons/Desktop/NSS/data analytics bootcamp/capstone/Data/crosswalk files/TRACT_ZIP_032016.xlsx')
tractzip17 = pd.read_excel('/Users/k.simmons/Desktop/NSS/data analytics bootcamp/capstone/Data/crosswalk files/TRACT_ZIP_032017.xlsx')
tractzip18 = pd.read_excel('/Users/k.simmons/Desktop/NSS/data analytics bootcamp/capstone/Data/crosswalk files/TRACT_ZIP_032018.xlsx')
tractzip19 = pd.read_excel('/Users/k.simmons/Desktop/NSS/data analytics bootcamp/capstone/Data/crosswalk files/TRACT_ZIP_032019.xlsx')
```

```python
# creating filter for census tracts that are only found in the disease data

tractmask = nashdisease2016['TractFIPS'].to_list()
```

```python
# filtering tractzip16 to only show the tracts that are found in the disease data

diseasetracts16 = tractzip16['TRACT'].isin(tractmask)
nashtractzip16 = tractzip16[diseasetracts16]
nashtractzip16.head(2)
```

```python
# now that tractzip16 contains only revelant tracts, filtering further to only include zips from ziplist

zipfilter16 = nashtractzip16['ZIP'].isin(ziplist)
nashtractzip16 = nashtractzip16[zipfilter16]
nashtractzip16.head(2)
```

```python
#dropping unnecessary columns

nashtractzip16 = nashtractzip16.drop(['RES_RATIO', 'BUS_RATIO', 'OTH_RATIO', 'TOT_RATIO'], axis=1)
nashtractzip16.head(2)
```

```python
#resetting index, and then setting tract number to index for ease of making dictionary using .to_dict()

nashtractzip16 = nashtractzip16.reset_index(drop=True)
nashtractzip16.set_index('TRACT', inplace=True)
nashtractzip16.head(2)
```

```python
tractzip16_dict = nashtractzip16.to_dict()
```
